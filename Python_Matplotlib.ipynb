# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = np.random.normal(size=500).cumsum() # random walk
b = np.random.normal(size=500).cumsum()
```

```python
tips = sns.load_dataset("tips")
```

## 2D

```python
plt.plot(a, label='A', color='#fabada')
plt.plot(b, label='B', color='#b3e3c3')
plt.xlabel('time', fontsize=10)
plt.ylabel('Value', fontsize=10)
plt.title('Evolution in time')
plt.grid(color='lightgray')
plt.legend()
plt.show()
```

Write `;` is the same as plt.show() in Jupyter.

```python
fig = plt.figure(figsize=(12, 3))
ax = plt.axes()
plt.plot(a, label='A', color='#fabada')
plt.plot(b, label='B', color='#b3e3c3')
plt.xlabel('time', fontsize=10)
plt.ylabel('Value', fontsize=10)
plt.title('Evolution in time')
plt.grid(color='lightgray')
plt.legend()
fig.savefig('Evolution.png')
plt.show()
```

```python
fig = plt.figure(figsize=(12, 3))
ax = plt.axes()
ax.plot(a, label='A', color='#fabada')
ax.plot(b, label='B', color='#b3e3c3')
ax.set_xlabel('time', fontsize=10)
ax.set_ylabel('Value', fontsize=10)
ax.set_title('Evolution in time')
ax.grid(color='lightgray')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2,4, figsize=(20,7))
ax[0,0].plot(a, label='A', color='#fabada')
ax[0,0].plot(b, label='B', color='#b3e3c3')
ax[0,1].plot(b, label='B', color='#b3e3c3')
ax[1,-2].scatter(x=tips.total_bill,y=tips.tip, s=10,color=['pink' if gender =='Female' else 'lightblue' for gender in tips.sex])
ax[0,-1].pie(tips.day.value_counts(), labels=tips.day.value_counts().index, colors=['pink','lightblue','#fabada','#b3e3c3'])
ax[1,-3].hist(a, color='#fabada')
ax[1,-4].grid(color='lightgrey')
plt.show()
```

## 3D

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter3D(xs=[-1,0,4],ys=[2,2,3],zs=[2,3,-1])
plt.show()
```

```python
x = np.linspace(-10,10,50)
y = np.linspace(-10,10,50)
```

```python
X, Y = np.meshgrid(x,y)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter3D(xs=X,ys=Y,zs=np.sin(np.sqrt(X**2+Y**2)), color='#b3e3c3',s=5)
plt.show()
```

```python
plt.colormaps()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot_surface(X=X,Y=Y,Z=np.sin(np.sqrt(X**2+Y**2)), cmap='plasma')
plt.show()
```

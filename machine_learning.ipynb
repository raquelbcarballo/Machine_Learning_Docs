# Machine Learning

Machine learning is a subfield of artificial intelligence (AI) that focuses on the development of algorithms and models that enable computers to learn and make predictions or decisions without being explicitly programmed.

## Model

A model is a mathematical function that receives certain values and return a prediction. The model depends on the data. Requires a procedure to configure it.

An **algorithm** is a set of step-by-step instructions or a well-defined sequence of operations that can be followed to create this model = **Training**.

After this training we need to **validate** this model.

`Data + Algotithm = Model (this procedure calls Training) --> Validation`

To refer to specific elements of a dataset that are used to train a predictive model there are:
* Features: also known as input variables or independent variables, are the measurable characteristics or attributes of the data points in a dataset. They represent the information that the model will use to make predictions or draw conclusions. Features can take various forms depending on the problem and the type of data being analyzed.

* Target: also referred to as the output variable or dependent variable, is the specific attribute or value that the machine learning model aims to predict or estimate based on the provided features. It is the variable that the model is trained to learn and make predictions on. The target variable is typically a known value in the training dataset, which allows the model to learn the relationship between the features and the target. Once trained, the model can then predict the target variable for new, unseen instances based on the learned patterns from the training data.

## Algorith

In machine learning, algorithms can be categorized into different types based on their characteristics and the way they learn from data.

* Supervised Learning: equire tagged or labeled data for training. In this approach, the training dataset consists of input features (also called predictors or independent variables) along with their corresponding output labels or target variables. The algorithm learns patterns and relationships between the features and labels in the training data to make predictions or classify new, unseen instances. Examples of supervised learning algorithms include decision trees, random forests, support vector machines (SVM), and neural networks.

  * Classification: the goal is to categorize or classify data points into different predefined classes or categories. The algorithm learns from labeled training data, where each data instance is associated with a known class label. The objective is then to build a model that can accurately assign the correct class label to new, unseen instances based on their features.

  * Regression: involves predicting continuous numerical values or quantities based on input features. In regression, the algorithm learns from labeled training data where each data instance has associated input features as well as a corresponding continuous target or output value. The objective is to build a model that can accurately estimate or predict the target value for new, unseen instances.

* No supervised Learning: do not require labeled data for training. Instead, they analyze the input data to discover patterns, relationships, or structures within it. These algorithms are used when the task is to explore the data and find hidden patterns or groupings. Common unsupervised learning techniques include clustering algorithms like k-means clustering, hierarchical clustering, and dimensionality reduction methods like principal component analysis (PCA) or t-distributed stochastic neighbor embedding (t-SNE). Unsupervised learning can be useful for tasks such as customer segmentation or anomaly detection.

  * Clustering: aims to group similar data points together based on their intrinsic characteristics or patterns. The goal is to discover inherent structures or clusters within a dataset without any prior knowledge of the class labels or categories. In clustering, the algorithm analyzes the input data and organizes it into clusters, where data points within the same cluster are more similar to each other compared to those in different clusters. Clustering algorithms group data points based on their proximity or similarity in terms of features or distance measures. Common clustering algorithms include k-means clustering, hierarchical clustering, and DBSCAN (Density-Based Spatial Clustering of Applications with Noise). Clustering can be used for various purposes such as customer segmentation, image segmentation, anomaly detection, and recommendation systems.

  * Dimensionality reduction: refers to the process of reducing the number of input features or variables while retaining important information. It is used to address the curse of dimensionality, where datasets with a large number of features can lead to computational inefficiency and overfitting. Dimensionality reduction techniques aim to transform the high-dimensional data into a lower-dimensional representation that still captures the essential information and structure of the original data.

   * Apriori (shopping list): a technique in data mining that discovers interesting relationships or associations between items in large datasets. The Apriori algorithm is specifically designed to identify frequent itemsets, which are sets of items that often co-occur together in the dataset.


* Reinforcement Learning: involve an agent that learns to make decisions or take actions in an environment to maximize a cumulative reward. The agent interacts with the environment and receives feedback in the form of rewards or penalties based on its actions. By exploring and exploiting different actions, the agent learns to optimize its decision-making process over time. Boosting algorithms, such as AdaBoost and Gradient Boosting, are not specific to reinforcement learning but rather belong to the supervised learning category. Boosting techniques create an ensemble model by combining multiple weak models iteratively, where each subsequent model focuses on the mistakes made by the previous models. The resulting ensemble is a stronger model that can make more accurate predictions.

### The perceptron

The perceptron is a basic unit of an artificial neural network (ANN) and a linear binary classifier. It was developed in the late 1950s by Frank Rosenblatt and is one of the earliest forms of machine learning algorithms.

The perceptron is designed to simulate the behavior of a single biological neuron. It takes one or more input signals, each multiplied by a corresponding weight. These weighted inputs are summed up, and a bias term is added to the sum. The resulting value is then passed through an activation function, typically a step function or a sigmoid function, to produce the output of the perceptron.

Mathematically, the output of a perceptron can be represented as follows: `output = activation_function(weighted_sum + bias)`

The learning rate, is a hyper-parameter used to govern the pace at which an algorithm updates or learns the values of a parameter estimate. In other words, the learning rate regulates the weights of our neural network concerning the loss gradient.

In the context of machine learning and deep learning, an **epoch** refers to one complete iteration through the entire dataset during the training phase. During an epoch, the model processes each training sample once, calculates the loss, and updates the model's parameters based on the chosen optimization algorithm. The number of epochs is a hyperparameter that determines how many times the model will train on the entire dataset. Choosing the appropriate number of epochs depends on factors such as the complexity of the problem, the size of the dataset, and the model's convergence behavior. It is often determined through experimentation and validation performance monitoring.

### Adaline

Adaptive Linear Neuron, is a type of artificial neural network (ANN) developed in the late 1950s and early 1960s by Bernard Widrow and Ted Hoff. It is a single-layer neural network that serves as a precursor to the more advanced multilayer perceptron (MLP).

The Adaline algorithm is primarily used for binary classification tasks, where the goal is to classify input data into one of two possible categories. It consists of a linear combination of the input features followed by an activation function, which determines the output of the network.

The main feature of Adaline is its adaptive learning rule, known as the Widrow-Hoff rule or the Delta rule. This rule adjusts the weights of the network in response to the difference between the predicted output and the desired output. The weight adjustments aim to minimize the error and improve the network's performance over time.

Unlike the perceptron, which uses a step function as its activation function, Adaline typically employs a linear activation function. This allows for real-valued outputs, which can be interpreted as continuous predictions rather than discrete classes. The linear activation function also facilitates the use of gradient descent algorithms for weight optimization.

Adaline has been used in various applications, including pattern recognition, signal processing, and adaptive filtering. Although it is a relatively simple neural network model compared to more modern architectures, it played a significant role in the development of the field and served as a foundation for more complex models that followed.

## Errors

In machine learning, an error function, also known as a loss function or cost function, is used to quantify the discrepancy between the predicted output of a model and the true output. The choice of an appropriate error function depends on the specific learning task and the nature of the data. The goal is to minimize the error function during the training process, which guides the model to make better predictions.

Here are some commonly used error functions in machine learning:

* Mean Squared Error (MSE): This is a popular error function used in regression tasks. It calculates the average squared difference between the predicted values and the true values. MSE gives higher penalties to larger errors, making it sensitive to outliers.

* Binary Cross-Entropy: Binary cross-entropy is commonly used in binary classification tasks. It measures the dissimilarity between the predicted probabilities and the true binary labels. It is defined as the average of the logarithmic loss across all samples.

* Categorical Cross-Entropy: Categorical cross-entropy is used in multi-class classification problems. It measures the dissimilarity between the predicted class probabilities and the true class labels. It is an extension of binary cross-entropy to handle multiple classes.

* Hinge Loss: Hinge loss is often used in support vector machines (SVMs) and binary classification tasks. It aims to maximize the margin between classes. Hinge loss penalizes misclassifications and encourages correct predictions that are confident but not excessively so.

* Mean Absolute Error (MAE): MAE is another error function used in regression tasks. It calculates the average absolute difference between the predicted values and the true values. MAE is less sensitive to outliers compared to MSE.

* Huber Loss: Huber loss is a combination of MSE and MAE. It behaves like MSE for smaller errors but switches to MAE for larger errors. Huber loss is more robust to outliers and can be used in regression tasks where outliers are present.

## Optimizers

Optimizers in machine learning are algorithms or methods that are used to adjust the parameters of a model during the training process in order to minimize the error or loss function. These optimization algorithms determine how the model's parameters are updated based on the computed gradients. Here are some commonly used optimizers in machine learning:

* Gradient Descent (GD): Gradient descent is a fundamental optimization algorithm used to update the parameters of a model. It calculates the gradients of the loss function with respect to the parameters and adjusts the parameters in the direction of steepest descent. There are different variants of gradient descent, including batch gradient descent, stochastic gradient descent (SGD), and mini-batch gradient descent.

<img src="./misc/gradient.png" width="500" height="300">

* Stochastic Gradient Descent (SGD): SGD is a variant of gradient descent that updates the model's parameters using only a single training example (or a small batch) at each iteration. It can be faster than batch gradient descent, especially for large datasets, as it updates the parameters more frequently. However, SGD can have higher variance due to the noisy updates.

* Adam: Adam (Adaptive Moment Estimation) is a popular optimization algorithm that combines the benefits of both AdaGrad and RMSProp. It maintains adaptive learning rates for each parameter and computes adaptive momentum. Adam is efficient, performs well on a wide range of problems, and is widely used in deep learning.

### Derivatives

In mathematics, a derivative represents the rate at which a function changes with respect to its input variables. It measures the slope or the steepness of a function at a particular point.

Formally, the derivative of a function f(x) at a point x = a is denoted as f'(a) or dy/dx|a (read as "the derivative of y with respect to x evaluated at a"). Geometrically, it represents the slope of the tangent line to the graph of the function at the point (a, f(a)).

The derivative is calculated using the concept of limits. For a continuous function, the derivative at a point a is defined as the limit of the difference quotient as the interval around a approaches zero:

In this formula, h is a small increment in the input variable. As h approaches zero, the quotient measures the average rate of change of the function over a smaller and smaller interval, ultimately giving the instantaneous rate of change at the point a.

When the derivative of a function is zero at a specific point, it means that the function is not changing its value at that point. In other words, the function has a stationary point or a critical point where the slope of the function is flat.

There are a few possibilities for what can happen when the derivative is zero:

Local maximum or minimum: If the derivative is zero at a point, and the derivative changes from positive to negative as we move from left to right around that point, it indicates that the function has a local maximum at that point. Conversely, if the derivative changes from negative to positive, it indicates a local minimum.

Inflection point: In some cases, the derivative can be zero at a point where the function has an inflection point. An inflection point is a point on the graph where the curve changes concavity. At an inflection point, the function is neither at a local maximum nor a minimum.

Plateau or flat region: Sometimes, the derivative can be zero at a point where the function exhibits a flat region or a plateau. In this case, the function remains constant in a certain interval around the point where the derivative is zero.

Endpoint or boundary: At the endpoints or boundaries of a function, the derivative can be zero. This occurs when the function does not exist or is not defined beyond those points.

Determining the nature of a point where the derivative is zero requires additional analysis, such as examining the second derivative or evaluating the function at that point. This analysis helps to classify the point as a maximum, minimum, inflection point, or flat region.

### Overfitting

Occurs when a machine learning model performs exceptionally well on the training data but fails to generalize well to unseen or new data. It happens when the model becomes too complex or too specialized to the training data, effectively memorizing noise or specific patterns that are not representative of the underlying true relationship.

Signs and consequences of overtraining:

* High training accuracy, low test accuracy: The model achieves very high accuracy or low error on the training data but performs poorly on the test or validation data. This indicates that the model is unable to generalize well beyond the training set.

* Overly complex model: Overtrained models tend to be overly complex with many parameters or features, often resulting in high variance. They capture noise or irrelevant details in the training data, making them less effective in capturing the true underlying patterns.

* Sensitivity to small changes: Overtrained models can be sensitive to small variations or noise in the input data, leading to unstable and unreliable predictions.

* Overconfidence in predictions: Overtrained models may exhibit excessive confidence in their predictions, even when they are incorrect. This can lead to poor decision-making or misleading results.

Preventing and mitigating overtraining:

* Cross-validation: Use techniques like k-fold cross-validation to assess the model's performance on multiple subsets of the data. This helps in detecting overtraining by evaluating how the model generalizes to different data partitions.

* Regularization: Apply regularization techniques such as L1 or L2 regularization, which introduce a penalty term to the loss function. This discourages excessive complexity in the model and helps prevent overfitting.

* Feature selection: Select relevant and informative features to reduce the dimensionality of the input space. Eliminating irrelevant or redundant features can help prevent overtraining.

* Early stopping: Monitor the model's performance on a validation set during training. Stop training when the performance on the validation set starts to degrade, indicating that further training may lead to overfitting.

* Data augmentation: Increase the size and diversity of the training data by applying data augmentation techniques, such as rotation, translation, or adding noise. This can help expose the model to a wider range of variations and reduce overtraining.

* Model simplicity: Consider using simpler models with fewer parameters or a smaller number of layers. Simple models are less prone to overtraining and often generalize better.

By implementing these strategies, you can reduce the likelihood of overtraining and improve the model's ability to generalize to unseen data, leading to better overall performance.

### Gini index
The Gini index is a measure of impurity used in decision tree algorithms, particularly when deciding the best attribute to split the data. It quantifies the degree of impurity or disorder of a node in a decision tree.

In the context of classification problems, the Gini index assesses the probability of misclassifying a randomly chosen element in a node if it were randomly labeled according to the class distribution of the node. It ranges from 0 to 1, where 0 indicates perfect purity (all elements belong to the same class) and 1 indicates maximum impurity (an equal distribution of elements across all classes).

Mathematically, the Gini index for a node "i" with "C" classes can be calculated as:

`Gini(i) = 1 - ∑(p(j)^2)`

where "p(j)" represents the probability of an element in the node belonging to class "j."

When splitting a node based on a particular attribute, the Gini index is used to evaluate the impurity reduction achieved by that split. The attribute with the lowest Gini index after the split is selected as the best attribute for the split, as it leads to the greatest reduction in impurity.

The Gini index is commonly used alongside other attribute selection measures, such as information gain and gain ratio, to determine the most appropriate attribute and split point during the construction of a decision tree.

### Enthropy
Entropy, in the context of machine learning and information theory, is a measure of impurity or uncertainty in a set of data. It is commonly used as an attribute selection measure in decision tree algorithms to determine the best attribute to split the data.

In the context of classification problems, entropy quantifies the degree of disorder or randomness in a set of class labels. It measures the average amount of information needed to identify the class of an element drawn from the set.

Mathematically, the entropy of a set "S" with respect to a set of classes "C" can be calculated as:

`Entropy(S) = - ∑ (p(c) * log2(p(c)))`

where "p(c)" represents the probability of an element in "S" belonging to class "c". The summation is taken over all classes in "C".

Entropy ranges from 0 to log2(C), where 0 indicates perfect purity (all elements belong to the same class) and log2(C) indicates maximum impurity (an equal distribution of elements across all classes).

When splitting a node based on a particular attribute, the entropy is used to evaluate the information gain achieved by that split. The attribute with the highest information gain or the largest reduction in entropy after the split is selected as the best attribute for the split.

By minimizing the entropy or maximizing the information gain, decision tree algorithms aim to find the attribute and split that provide the most effective separation of classes, leading to more accurate and informative decision trees.

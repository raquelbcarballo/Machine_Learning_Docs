```python
sns.set_palette('pastel')
```

```python
sns.relplot(data = tips, x='total_bill',y='tip', hue='sex',style='day', size='size',aspect=3)
plt.show()
```

```python
fig, ax = plt.subplots(2,3, figsize=(20,7))
sns.scatterplot(data = tips, x='total_bill',y='tip', hue='sex',ax=ax[0,0])
sns.histplot(data=tips,x='total_bill', ax=ax[0,1])
sns.boxplot(data=tips,x='total_bill', ax=ax[0,2])
sns.barplot(data=tips,y='total_bill',x='day', ax=ax[1,0])
sns.lineplot(data=tips,y='total_bill',x='day', ax=ax[1,1])
sns.violinplot(data=tips,y='total_bill',x='day', ax=ax[1,2])
plt.show()
```
